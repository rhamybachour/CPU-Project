#=============== Simulações ==================

##Simulação com 2 processadores usando SingleQueue.

```python
###### Declaração de Variáveis ######
single_lista_tempos = []
single_lista_chegada_saida_tarefa = []
single_lista_tempo_livre_processador = []
single_lista_tempo_extra_tarefa = []

###### Simulação ######
for x in range(10):
    tempo = single_queue_simulation(1000, lambda i: random.lognormvariate(2, 1.0), 1, 0.25, 2, 0.1, 0.3)
    single_lista_tempos.append(tempo[0])
    single_lista_chegada_saida_tarefa += tempo[1]
    single_lista_tempo_livre_processador += tempo[2]

###### Manipulação dos Dados Obtidos ######
for i in range(10000):
    single_lista_tempo_extra_tarefa.append(single_lista_chegada_saida_tarefa[i][1] - single_lista_chegada_saida_tarefa[i][0] - single_lista_chegada_saida_tarefa[i][2])    

single_array_tempos = np.array(single_lista_tempos)
print('O tempo médio foi', single_array_tempos.mean(),'unidades de tempo')
print('O desvio padrão foi', single_array_tempos.std())


```

##Simulação com 2 processadores usando MultiQueue.

```python
###### Declaração de Variáveis ######
multi_lista_tempos = []
multi_lista_chegada_saida_tarefa = []
multi_lista_tempo_livre_processador = []
multi_lista_tempo_extra_tarefa = []

###### Simulação ######
for x in range(10):
    tempo = multi_queue_simulation(1000, lambda i: random.lognormvariate(2, 1.0), 1, 0.25, 2, 0.1, 0.3)
    multi_lista_tempos.append(tempo[0])
    multi_lista_chegada_saida_tarefa += tempo[1]
    multi_lista_tempo_livre_processador += tempo[2]

###### Manipulação dos Dados Obtidos ######
for i in range(10000):
    multi_lista_tempo_extra_tarefa.append(multi_lista_chegada_saida_tarefa[i][1] - multi_lista_chegada_saida_tarefa[i][0] - multi_lista_chegada_saida_tarefa[i][2])

multi_array_tempos = np.array(multi_lista_tempos)
print('O tempo médio foi', multi_array_tempos.mean(),'unidades de tempo')
print('O desvio padrão foi', multi_array_tempos.std())
```

##Gráfico de tempo extra das tarefas para os dois tipos de simulações.

```python
###### Plot ######
y, retangulo = np.histogram(single_lista_tempo_extra_tarefa)
centro_retangulo = 0.5*(retangulo[1:] + retangulo[:-1])
single_plot = plt.plot(centro_retangulo, y, '-')

y, retangulo = np.histogram(multi_lista_tempo_extra_tarefa)
centro_retangulo = 0.5*(retangulo[1:] + retangulo[:-1])
multi_plot = plt.plot(centro_retangulo, y, '-')

###### Configurações ######
plt.grid(True)
plt.title('Gráfico 1 - Tempo Extra das Tarefas', fontsize=16)
plt.xlabel('Unidades de Tempo', fontsize=13)
plt.ylabel('Quantidade de Vezes', fontsize=13)
single_setup = plt.setp(single_plot, color='r', linewidth=2.0, label='Single Simulation')
multi_setup = plt.setp(multi_plot, color='b', linewidth=2.0, label='Multi Simulation')
plot_setup = plt.legend(loc='upper right')
```

##Gráfico de tempo livre dos processadores para os dois tipos de simulações.

```python
###### Plot ######
y, retangulo = np.histogram(single_lista_tempo_livre_processador)
centro_retangulo = 0.5*(retangulo[1:] + retangulo[:-1])
single_plot = plt.plot(centro_retangulo, y, '-')

y, retangulo = np.histogram(multi_lista_tempo_livre_processador)
centro_retangulo = 0.5*(retangulo[1:] + retangulo[:-1])
multi_plot = plt.plot(centro_retangulo, y, '-')

###### Configurações ######
plt.grid(True)
plt.title('Gráfico 2 - Tempo Ociosos dos Processadores', fontsize=16)
plt.xlabel('Unidades de Tempo', fontsize=13)
plt.ylabel('Quantidade de Vezes', fontsize=13)
single_setup = plt.setp(single_plot, color='r', linewidth=2.0, label='Single Simulation')
multi_setup = plt.setp(multi_plot, color='b', linewidth=2.0, label='Multi Simulation')
plot_setup = plt.legend(loc='upper right')
```

Comparando os resultados obtidos e observandos os Gráfico 1 e 2, observamos que no geral, a Multi Simulation foi eficaz, pois fez os mesmos processamentos que a Single Simulation, porém mais rapido, mesmo com maiores tempos ociosos dos processadores.
